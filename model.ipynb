```python
ultralytics.SETTINGS['loggers'] = ['wandb']
ultralytics.SETTINGS['wandb'] = {'project': 'SeoulTechML', 'name': 'yolo11m'}
```

```python
model = ultralytics.YOLO('yolo11m.pt')
```

```python
results = model.train(data="yolo.yaml", epochs=3, save=True, device=0)
```
